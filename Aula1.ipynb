```python
import time
import random
```

```python
def insertionSort(arr):
    for i in range(1, len(arr)):
        j = i
        while j > 0 and arr[j-1] > arr[j]:
            arr[j], arr[j-1] = arr[j-1], arr[j]
            j -= 1
```

```python
# Testando o ponto de ruptura
sizes = [10, 20, 50, 100, 500, 1000, 5000, 10000]
for n in sizes:
    test_data = [random.randint(0, 10000) for _ in range(n)]

    # Measure time for insertionSort
    data_for_insertion = test_data.copy() # copia dos dados para não modificar o array original
    start_time_insertion = time.perf_counter()
    insertionSort(data_for_insertion) # nossa funçao sorted aqui
    end_time_insertion = time.perf_counter()
    t_insertion = end_time_insertion - start_time_insertion

    # Tempo do sorted()
    data_for_merge = test_data.copy() # copia de novo
    start_time_merge = time.perf_counter()
    sorted(data_for_merge) # a built-in sorted do python
    end_time_merge = time.perf_counter()
    t_merge = end_time_merge - start_time_merge

    print(f"N={n} | Insertion: {t_insertion:.5f}s | Sorted (Python): {t_merge:.5f}s")
```

Aqui podemos notar que a função insertion sort tem um crescimento O(n^2), ou seja apresenta crescimento quadrático. Isso significa que, ao dobrar o tamanho da entrada, o tempo tende a quadruplicar:

exemplo:
N=5000 | Insertion: 1.04311s
N=10000 | Insertion: 4.17106s

Já a sorted que é O(nlogn) tem um crescimento logaritmo e por isso é mais rapida ao ordenar listas grandes.

Já o ponto de ruptura, momento em que o algorítmo O(n^2) passa a ter um custo tão alto que não se pode mais compara com o algorítmo O(nlogn). Até N = 100 a diferença é de de milesimos de segundos, mas mais que isso já se pode ver diferença significativa de tempo
