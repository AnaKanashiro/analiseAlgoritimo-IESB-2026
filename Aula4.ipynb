F(n)= 2F(n-1)+n^2

```python
def aula4(n):
  if n == 1:
    return 2
  if n > 1:
    return(2*aula4(n-1)+n*n)
```

```python
n = int(input("Digite um número inteiro: "))
print("exericio de recurssao:", aula4(n))
```
