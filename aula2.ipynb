```python
import time
```

```python
def factorial(n):
    if n < 0:
        return "Invalid input"
    if n == 0 or n == 1:
        return 1

    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
```

```python
numArr = [10, 100, 500, 750]

for n in numArr:
    start_time = time.perf_counter()  # mais preciso para benchmarking
    factorial_n = factorial(n)
    end_time = time.perf_counter()

    execution_duration = end_time - start_time

    print(f"Factorial of {n}: {factorial_n}")
    print(f"Execution time: {execution_duration:.8f} seconds")
```

A versão recursiva do fatorial tem tempo de execução O(n), pois realiza uma chamada para cada valor até chegar ao caso base. O problema principal está no uso de memória: como cada chamada fica armazenada na pilha, a função também ocupa O(n) espaço. Isso significa que, ao aumentar n, cresce não só o tempo, mas também o consumo de RAM, tornando a abordagem arriscada para valores muito grandes devido ao limite de recursão.
